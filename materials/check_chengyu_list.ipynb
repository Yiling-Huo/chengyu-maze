```python
import csv
with open('chengyu-list.csv', 'r', encoding='utf-8') as file:
    cr = csv.reader(file)
    content = [line for line in cr]
    check = []
    dup = []
    for line in content:
        if line[0] in check:
            dup.append(line[0])
        else:
            check.append(line[0])
print(dup)
```
